# TODO 2

PCA  .

## 
1. normalize  
2.  numpy    
    - ```numpy.mean()```
    - ```numpy.linalg.eig()```
    - ```numpy.dot()```
    - ```numpy.cov()```

````python
def student_pca(X, n_components=2):
    pca_results = None
    # TODO -->
    pca_X=np.transpose(X)
    pca_cov=np.cov(pca_X)
    #print(pca_cov.shape,X.shape)
    eigenValues,eigenVectors=np.linalg.eig(pca_cov)
    idx = eigenValues.argsort()[::-1]   
    eigenValues = eigenValues[idx]
    eigenVectors = eigenVectors[:,idx]
    #print(eigenVectors.shape)
    #print(X.shape,(np.transpose(eigenVectors[:,: n_components]).shape))
    pca_results=np.transpose(np.dot(np.transpose(eigenVectors[:,: n_components]),pca_X-pca_X.mean()))
    # <--
    return pca_results

pca_student_score = student_pca(X, n_components=2)
````

````python
show2D(pca_student_score, Y, target_names)
````
